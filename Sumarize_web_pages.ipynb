```python

# se llaman a las librerias que se usaran 
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
import ollama

MODEL = "gemma3"

```

```python

#creamos la clase que extraera la informacion de la pagina web
# usamos beautifulsoup para acceder a la informacionde la pagina y obtener el titulo y texto que la compone

class Website:
   

    def __init__(self, url):
        
        self.url = url
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No tiene título"
        #excluimos elementos irrelevantes de la pagina web
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True)
```

```python
#creamos un objeto el cual hace referencia a la pagina web que vamos a consultar 
wiki = Website("https://es.wikipedia.org/wiki/Wikipedia:Portada")
print(wiki.title)
print(wiki.text)
```

```python
#creamos una funcion para el promt que le pasaremos a nuestro modelo y asi especificarle de forma clara que debe hacer 
#para resumir la pagina web que le pasamos 

def mensaje(website):
    user_prompt = f"Estás viendo un sitio web titulado {website.title}"
    user_prompt += "\nEl contenido de este sitio web es el siguiente; \
    proporciona un breve resumen de este sitio web en formato Markdown. \
    Si incluye noticias, productos o anuncios, resúmelos también.\n\n"
    user_prompt += website.text
    return user_prompt
```

```python
#definimos el mensaje que le pasaremos a ollama 

def messages_for(website):
    return [
        
        {"role": "user", "content": mensaje(website)}
    ]
```

```python
#ejecutamos el codigo de ollama para hacer nuestra consulta a gemma3

response = ollama.chat(model=MODEL, messages=messages_for(wiki))
print(display(Markdown(response['message']['content'])))  #usamos el display y el markdown para que se imprima en este formato
```
